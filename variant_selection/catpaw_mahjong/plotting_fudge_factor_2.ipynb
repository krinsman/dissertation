```python
import numpy as np
import matplotlib.pyplot as plt
```

A better definition of $\Phi$ from previous notebook, corresponding to more exact/precise Stirling's approximation. A consequence is that it is no longer defined at $C$, i.e. only defined on $[0, C)$, even though the one in the other notebook could be extended to be defined on $[0,C]$.

But since the left term approaches $\frac{1}{2}$ instead of $0$ now as $x \to C$, the potential singularity isn't quenched (because $1/(C-x + \frac{1}{2}) \to 2$ as $x \to C$, whereas $1/(C-x) \to \infty$ as $x \to C$, and evidently $1/(C-x)$ blows up to infinity faster than $\log(C-x)$ blows "down" to negative infinity. So before L'Hospital's rule gave that the limit was defined and equal to $0$, whereas now there just simply cannot be any limit.

```python
# x = sampled number of cells (of strain), C = total number of cells (of strain)
phi = lambda x, C:  (C - x + 0.5)*( np.log(C) - np.log(C - x)  )
```

```python
x_values = np.arange(start=0, stop=100, step=0.001)
y_values = phi(x_values, 100)
```

```python
plt.title('$\Phi(x,100)$')
plt.xlabel('$x$')
plt.plot(x_values, y_values)
plt.show()
```

Is there a little kink at the right endpoint which I spy with my little eye?

```python
plt.title('$\Phi(x,100)$')
plt.xlabel('$x$')
plt.plot(x_values, y_values)
plt.show()
```

OK, so apparently the singularity only happens _very_ close to $C$, but it still clearly happens at $C$, right? So I rest my case.

```python
x_values = np.arange(start=99.99999, stop=100, step=0.00000001)
y_values = phi(x_values, 100)
```

```python
plt.title('$\Phi(x,100)$')
plt.xlabel('$x$')
plt.plot(x_values, y_values)
plt.show()
```

Hard to get a good plot showing it of course, since it evidently appears to happen _so_ close to the end of the domain of definition. Apparently if you blinked you'd miss it. But L'Hospital's Rule fails, so mathematically a singularity has to exist, again even if it happens so close to the right endpoint of the domain that numerical precision prevents it from being plotted clearly. In any case the trend is pretty clear, the value nearest to the singularity can be seen to rise higher and higher each time another "zoom in" occurs.
