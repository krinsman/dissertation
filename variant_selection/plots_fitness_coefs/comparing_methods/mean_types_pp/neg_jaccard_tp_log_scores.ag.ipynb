```python
# Parameters
base_result_filename = "fitness_coefs/analyzed_fitness_coef_results_pp.npz"
results_dir = "../.."
kde_bounds = None
right_y_title = (
    "Fraction of Random Guesses with Strictly Worse Negative Jaccard Similarity"
)
transformation_name = "butterfly"
start_indices = [0]
metric_name = "neg_jaccard_tp"
y_title = "Log Scores of Fraction of Random Guesses with Strictly Worse Negative Jaccard Similarity"
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from plot_utils import fitnessVariantsPlotter as Plotter
```

```python
transformations = {'identity': (lambda x: x), 'butterfly': (lambda x: -np.log(1. - x)/np.log(10))}
transformation = transformations[transformation_name]
```

```python
plotter = Plotter(base_result_filename, results_dir, kde_bounds)
```

```python
performance = plotter.load_performance_results(metric_name)
performance = transformation(performance)
```

# Method variants side by side

### Results for Averages of Normalized

```python
for start_index in start_indices:
    ax = plotter.methods_side_by_side(performance, avg_type="normalized", 
                                                    y_title=y_title, start_index=start_index)
    if transformation_name == 'butterfly':
        plotter.annotate_important_log_scores(ax, right_y_title)
```

### Results for Averages of Unnormalized

```python
for start_index in start_indices:
    ax = plotter.methods_side_by_side(performance, avg_type="unnormalized", 
                                                    y_title=y_title, start_index=start_index)
    if transformation_name == 'butterfly':
        plotter.annotate_important_log_scores(ax, right_y_title)
```
