```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
npzfile = np.load('brownian_sample_path.npz')
brownian = npzfile['brownian']
x = npzfile['x']
```

```python
fig, ax = plt.subplots(1,1, figsize=(30,20))
plt.plot(x, brownian)
plt.fill_between(x, brownian, 0, where=brownian > 0, color='green', alpha=0.7, label='Total Number of Cells that Ever Formed')
plt.fill_between(x, 0, brownian, where=brownian < 0, color='red', alpha=0.7, label='Total Number of Cells that Ever Died')
plt.axhline()
plt.xlim((0,0.001))
plt.legend(fontsize=36)
plt.xticks([])
plt.xlabel('Time', fontsize=72)
plt.yticks([])
plt.ylabel('Growth Rate', fontsize=72)
plt.savefig('cell_number_integration.png')
plt.show()
```
