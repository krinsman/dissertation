```python
from scipy.stats import chi2
xvals = np.arange(3500,4500)
yvals = chi2.pdf(xvals, df=4187)
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(15,9))

sns.lineplot(x=xvals, y=yvals, color='gray', ax=ax)
ax.fill_between(xvals, yvals, 0, alpha=0.3, color='gray')
#sns.kdeplot(cnbdm_100_divergences, fill=True, alpha=0.7, color=sns.color_palette("crest")[0], ax=ax)
#sns.histplot(cnbdm_100_divergences, ax=ax, alpha=0.7, stat="density", color=sns.color_palette("crest")[0])
#sns.kdeplot(cpdm_100_divergences, fill=True, alpha=0.7,color=sns.color_palette("flare")[0],ax=ax)
#sns.histplot(cpdm_100_divergences, ax=ax, alpha=0.7, stat="density", color=sns.color_palette("flare")[0])

sns.kdeplot(categorical_divergences, ax=ax, fill=True, alpha=0.7)
sns.histplot(categorical_divergences, ax=ax, alpha=0.7, stat="density")

ax.legend(['_nolegend_', '$\chi^2(_{91}C_2 + 91 + 2 - 1)$', 
           'CTPMHg'])
#           'CNBDM, $\zeta=100$', 'CPM, $\zeta=100$'])

ax.set_ylabel("Density", fontsize=18)
ax.set_xlabel("Pearson $\chi^2$ Divergence from Counts Expected Under CPM", fontsize=18)


#ax.legend([ '_nolegend_', '$\chi^2(_{91}C_2 + 91 + 2 - 1)$',  'CNBDM, $\zeta=100$', 'CPDM, $\zeta=100$'])
ax.set_title('Pearson $\chi^2$ Divergence from Counts Expected under CPM', fontsize=24)
```
