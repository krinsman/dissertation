```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from plot_utils.reflected_kde import initialize_reflected_kde
initialize_reflected_kde(kde_bounds=(0,1))

from plot_utils import cool_savefig
```

```python
hNBDM_100 = np.load('CNBDM_100/monte_carlo_pvals.npz')
hPoDM_100 = np.load('CPDM_100/monte_carlo_pvals.npz')
hTPMH = np.load('catpaw_mahjong/monte_carlo_pvals.npz')
```

```python
hTPMH.files
```

```python
hNBDM_100_pvals = hNBDM_100['monte_carlo_pvals']
hPoDM_100_pvals = hPoDM_100['monte_carlo_pvals']
hTPMH_pvals = hTPMH['monte_carlo_pvals']
```

## hTPMH

```python
with sns.axes_style("darkgrid"):
    plt.figure(figsize=(12, 8))
    ax = sns.kdeplot(hTPMH_pvals, fill=True, alpha=0.7)
    ax.set_ylabel("Estimated Probability Density", fontsize=20)
    ax.set_xlabel("Monte Carlo $p$-Values for Categorical Divergence Statistic", fontsize=20)
    sns.histplot(hTPMH_pvals, stat="density", bins=10, ax=ax)
    ax.set_xlim((0,1))
    ax.tick_params(labelsize=14)
    title = 'hTPMH - Goodness of Fit to hPoMu'
    ax.set_title(title, fontsize=24)
    cool_savefig(title)
```

## hPoDM $\zeta_C = 100$

```python
with sns.axes_style("darkgrid"):
    plt.figure(figsize=(12, 8))
    ax = sns.kdeplot(hPoDM_100_pvals, fill=True, alpha=0.7, color=sns.color_palette("flare")[0])
    ax.set_ylabel("Estimated Probability Density", fontsize=20)
    ax.set_xlabel("Monte Carlo $p$-Values for Categorical Divergence Statistic", fontsize=20)
    sns.histplot(hPoDM_100_pvals, stat="density", bins=10, color=sns.color_palette("flare")[0], ax=ax)
    ax.set_xlim((0,1))
    ax.tick_params(labelsize=14)
    title
    ax.set_title(title, fontsize=24)
    cool_savefig('hPoDM ($\zeta_C = 100$) - Goodness of Fit to hPoMu')
```

## hNBDM $\zeta = 100$

```python
with sns.axes_style("darkgrid"):
    plt.figure(figsize=(12, 8))
    ax = sns.kdeplot(hNBDM_100_pvals, fill=True, alpha=0.7, color=sns.color_palette("crest")[0])
    ax.set_ylabel("Estimated Probability Density", fontsize=20)
    ax.set_xlabel("Monte Carlo $p$-Values for Categorical Divergence Statistic", fontsize=20)
    sns.histplot(hNBDM_100_pvals, stat="density", bins=10, color=sns.color_palette("crest")[0], ax=ax)
    ax.set_xlim((0,1))
    ax.tick_params(labelsize=14)
    title = 'hNBDM ($\zeta = 100$) - Goodness of Fit to hPoMu'
    ax.set_title(title, fontsize=24)
    cool_savefig('hNBDM ($\zeta = 100$) - Goodness of Fit to hPoMu')
```
