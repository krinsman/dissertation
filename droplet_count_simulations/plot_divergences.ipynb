```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import matplotlib.ticker as mtick

def x_fmt(x, pos):
    return '$10^{{{}}}$'.format(np.round(x,2))#.replace('.0','').replace('.5', '')

from plot_utils import cool_savefig
```

```python
hTPMH = np.load('catpaw_mahjong/categorical_divergences_and_chi2_approx_pvals.npz')
hTPMH_divergences = hTPMH['pearson_categorical_divergences']
```

## hPoDM

```python
hExhPoDM = np.load('CECPDM/categorical_divergences_and_chi2_approx_pvals.npz')
hPoDM_1 = np.load('CPDM/categorical_divergences_and_chi2_approx_pvals.npz')
hPoDM_100 = np.load('CPDM_100/categorical_divergences_and_chi2_approx_pvals.npz')
```

```python
hExhPoDM_divergences = hExhPoDM['pearson_categorical_divergences']
hPoDM_1_divergences = hPoDM_1['pearson_categorical_divergences']
hPoDM_100_divergences = hPoDM_100['pearson_categorical_divergences']
```

```python
from scipy.stats import chi2
xvals = np.arange(3500,4500)
yvals = chi2.pdf(xvals, df=4187)
```

```python
sns.color_palette("flare")
```

```python
with sns.axes_style("darkgrid"):
    fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(15,9))

    sns.lineplot(x=xvals, y=yvals, color='gray', ax=ax, alpha=0.3)
    ax.fill_between(xvals, yvals, 0, alpha=0.3, color='gray')
    sns.kdeplot(hPoDM_100_divergences, fill=True, alpha=0.7, color=sns.color_palette("flare")[0], ax=ax)
    sns.kdeplot(hPoDM_1_divergences, fill=True, alpha=0.7,color=sns.color_palette("flare")[1],ax=ax)

    ax.set_ylabel("Probability Density", fontsize=20)
    ax.set_xlabel("Pearson Categorical Divergence from Counts Expected Under hPoMu", fontsize=20)
    ax.tick_params(axis='x', labelsize=14)

    sns.kdeplot(hExhPoDM_divergences, fill=True, alpha=0.7, color=sns.color_palette("flare")[3],ax=ax)

    ax.legend([ '_nolegend_', '$\chi^2(_{91}C_2 + 91 + 2 - 1)$',  
               '$\zeta_C=100$', '$\zeta_C=1$', '$\zeta_C \sim \mathrm{Exp}(1)$'], fontsize=18)
    ax.set_xlim((0,60000))
    
    title = 'hPoDM Distributions - Comparison to Homogeneous Model'
    ax.set_title(title, fontsize=24)
    cool_savefig(title)
```
